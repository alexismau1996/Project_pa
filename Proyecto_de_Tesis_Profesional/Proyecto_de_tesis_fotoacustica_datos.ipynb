```python
import sympy as sp
import numpy as np
import math
import matplotlib.pyplot as plt
import cmath
import time
import pandas as pd 
def sinh(x):
    return np.sinh(x)
def cosh(x):
    return np.cosh(x)
def exp(x):
    return np.exp(x)
def expi(x):
    return ((exp(x.real))*(complex(math.cos(x.imag),math.sin(x.imag))))
def coshi(x):
    re=((cosh(x.real))*(math.cos(x.imag)))
    im=((sinh(x.real))*(math.sin(x.imag)))
    return complex(re,im)
def sinhi(x):
    re=((sinh(x.real))*(math.cos(x.imag)))
    im=((cosh(x.real))*(math.sin(x.imag)))
    return complex(re,im)
def fase(x):
    n=len(x)
    fases=np.zeros(n)
    for i in range(0,n):
        if x[i].imag>0:
            fases[i]=(180/math.pi)*cmath.phase(x[i])-360
        else:
            fases[i]=(180/math.pi)*cmath.phase(x[i])
    return fases+360
```

```python

I=10
k=150
e=1.88*1.60218e-19
eg=1.11*1.60218e-19
Dp=1.2e-3
t=5e-6
beta=2.58e5
D=0.9e-4
ce=712
R=0.35
s1=2
s2=24
l=950e-6
lc=5e-2
To=298
gamma=1.395
alpha=2.6e-06
po=101325
Rs=4e-03
kg=0.02624

Dg=22e-6
dn=-9e-31
def computation(z,f,D1):
    w=2*math.pi*f
    Lp=(Dp*t)**0.5
    den=complex(1,w*t)
    L=Lp/((den)**0.5)
    vd=Dp/L
    vb=beta*Dp
    A=(beta*I)/(2*e*Dp*((beta**2) - (1/L**2)))
    y=(vd+s1)*(vd+s2)*expi(l/L)-(vd-s1)*(vd-s2)*expi(-l/L)
    n1=(vd-s2)*(vb+s1)*expi(-l/L)-(vd+s1)*(vb-s2)*exp(-beta*l)
    n2=(vd+s2)*(vb+s1)*expi(l/L)-(vd-s1)*(vb-s2)*exp(-beta*l)
    A1=A*n1/y
    A2=A*n2/y
    deltanp=A1*expi(z/L)+A2*expi(-z/L)-A*exp(-beta*z)
    mu=(D1/(math.pi*f))**0.5
    sigma=(1/mu)*(complex(1,1))
    mug=(Dg/(math.pi*f))**0.5
    sigmag=(1/mu)*(complex(1,1))
    r=beta/sigma
    c=(L*sigma)**(-1)
    C1=(I*beta*(e-eg))/(2*k*e*(beta**2 - sigma**2))
    g=0
    numterm=(g+r)*(1-g)*expi(sigma*(z-l))+(g+r)*(1+g)*expi(-sigma*(z-l))+(g-r)*exp(-beta*l)*((1+g)*expi(sigma*z)+(1-g)*expi(-sigma*z))
    denterm=(1+g)*(1+g)*expi(sigma*l)-(1-g)*(1-g)*expi(-sigma*l)
    Tterm=C1*(numterm/denterm -exp(-beta*z))
    delta0=A1+A2-A
    deltal=A1*expi(l/L)+A2*expi(-l/L)-A*exp(-beta*l)
    numtsr=s1*delta0*(coshi(sigma*(z-l))-g*sinhi(sigma*(z-l)))+s2*deltal*(coshi(sigma*z)+g*sinhi(sigma*z))
    Tsr=(2*eg/(k*sigma))*(numtsr/denterm)
    V1=((1/(r**2 -1))-(1+(delta0/A))/(c**2 -1))
    V2=((exp(-beta*l)+(deltal/A))/(c**2 -1))-((exp(-beta*l))/(r**2 -1))
    numv3=(1/A)*(deltal-delta0*coshi(l/L))-coshi(l/L)+exp(-beta*l)
    V3=-(c*numv3)/((c**2 -1)*sinhi(l/L))-(r/(r**2 -1))
    numv4=(1/A)*(deltal*coshi(l/L)-delta0)-1+(coshi(l/L))*exp(-beta*l)
    V4=(c*numv4)/((c**2 -1)*sinhi(l/L))+((r*exp(-beta*l))/(r**2 -1))
    com1=(V3+g*V1)*(1-g)*expi(-sigma*l)+(V4+g*V2)*(1+g)
    com2=(V3-g*V1)*(1+g)*expi(sigma*l)+(V4+g*V2)*(1-g)
    Tbr=(A*eg/(k*t*(sigma**2)))*(((com1*expi(sigma*z)+com2*expi(-sigma*z))/denterm) -(1/(c**2 -1))*((deltanp/A) +((r**2-c**2)/(r**2 -1))*exp(-beta*z)))
    return deltanp,Tterm,Tsr,Tbr
```

```python
def intnum(func,lower_limit, upper_limit, interval_limit ): 
    interval_size = ((upper_limit - lower_limit) / interval_limit) 
    sum = func[0] + func[len(func)-1]; 
    for i in range(1, interval_limit ): 
        if (i % 3 == 0): 
            sum = sum + 2 * func[i] 
        else: 
            sum = sum + 3 * func[i]
    return ((( 3 * interval_size) / 8 ) * sum ) 
```

```python

numpuntos=50
f=np.linspace(1,2e4+1,num=numpuntos,endpoint=False)
q=1j*f
fe_low1=15
fe_low2=25
fe_high1=14.9e3
fe_high2=9.5e3
fa_high1=14.7e3
fa_high2=9.4e3
damp_hf1=0.08*2*math.pi*fe_high1
damp_hf2=0.45*2*math.pi*fe_high2
HE_LF=-((f/fe_low1)/(1+q/fe_low1))*((f/fe_low2)/(1+q/fe_low2))
HE_HF=(1/(1+q/fe_high1))*(1/(1+q/fe_high2))
HA_HF=(((2*np.pi*fa_high1)**2)/((2*np.pi*fa_high1)**2+q*damp_hf1-f**2))*((((2*np.pi*fa_high2)**2)/((2*np.pi*fa_high2)**2+q*damp_hf2-f**2)))
H_EA=HE_LF+HE_HF+HA_HF
```

```python

def delta_p(f,l1,D1,alpha1):
    zdom=np.linspace(0,l1,num=901,endpoint=False)
    m=len(zdom)
    q=len(f)
    zl2=zdom-0.5*l1*np.ones(m)
    funts=np.zeros([m,q],dtype=complex)
    funtsdel=np.zeros([m,q],dtype=complex)
    deltanp=np.zeros([m,q],dtype=complex)
    Tterm=np.zeros([m,q],dtype=complex)
    Tsr=np.zeros([m,q],dtype=complex)
    Tbr=np.zeros([m,q],dtype=complex)
    delta_pa_te=np.zeros(q,dtype=complex)
    delta_pa_td=np.zeros(q,dtype=complex)
    delta_pa_pe=np.zeros(q,dtype=complex)
    for i in range(0,m):
        for j in range(0,q):
            [deltanp[i][j],Tterm[i][j],Tsr[i][j],Tbr[i][j]]=computation(zdom[i],f[j],D1)
    Tss=Tterm+Tsr+Tbr
    for k in range(0,q):
        funts[:,k]=(zl2)*(Tss[:,k])
        funtsdel[:,k]=(zl2)*(deltanp[:,k])
        delta_pa_te[k]=3*alpha1*gamma*po*((Rs**4)/((0.01*(2e-3)**2)*l**3))*intnum(funts[:,k],0,l1,m)
        delta_pa_pe[k]=3*dn*gamma*po*((Rs**4)/((0.01*(2e-3)**2)*l**3))*intnum(funtsdel[:,k],0,l1,m)
        delta_pa_td[k]=((1/To)*((gamma*po*(Dg**(0.5))))/((2*math.pi*f[k])**0.5))*Tss[m-1,k]/(((complex(0,1))**0.5)*0.01)    
    delta_tot=np.zeros(q,dtype=complex)
    delta_tot=delta_pa_te+delta_pa_pe+delta_pa_td
    return delta_tot
```

```python

start=time.time()
numdata=250
lv=1e-6*np.linspace(550,650,num=numdata,endpoint=False)
Dv=1e-5*np.linspace(9.3,9.5,num=numdata,endpoint=False)
alphav=1e-6*np.linspace(2.85,2.92,num=numdata,endpoint=False)
del_result1_amp=np.zeros((len(lv),len(f)))
del_result1_fas=np.zeros((len(lv),len(f)))
for i in range(0,len(lv)):
    del_result1_amp[i]=20*np.log10(np.abs(delta_p(f,lv[i],Dv[i],alphav[i])*H_EA))
    del_result1_fas[i]=fase(delta_p(f,lv[i],Dv[i],alphav[i])*H_EA)
signal_data=np.concatenate([del_result1_amp,del_result1_fas,1e6*lv.reshape(numdata,1),1e5*Dv.reshape(numdata,1),1e6*alphav.reshape(numdata,1)],axis=1)
lista= []
for i in range(1,numpuntos+1):
    lista.append("Pt_Ampl" + str(i))
for i in range(1,numpuntos+1):
    lista.append("Pt_phase" + str(i))
lista.append(r'l($\mu m$)')
lista.append(r'$D_{T}(10^{-5} m^{2}/s$)')
lista.append(r'$\alpha(K^{-1})$')
df_datos=pd.DataFrame(signal_data,columns=lista)
df_datos.to_excel (r'export_dataframe_2.xlsx', index = False, header=True)
end=time.time()
print(end-start)
df_datos
```
